#### Clean race census data in preparation for machine learning model

```python
# Read in demographic data race file
race_df = pd.read_csv("../CSV_Outputs/Combined_Census_Race_Data.csv",)

# Display sample data
race_df.head(5)
```

```python
# Copy the DataFrame and assign it a different name for 2008
race_2008_df = race_df.copy()
```

```python
# Filter columns
columns_to_keep_2008 = ['state'] + [col for col in race_2008_df.columns if col.endswith('_2008')]
race_2008_df = race_2008_df[columns_to_keep_2008]

race_2008_df.head()
```

```python
# Create a dictionary for renaming columns
new_column_names_2008 = {col: col.replace('_2008', '') for col in race_2008_df.columns if '_2008' in col}

# Rename columns
race_2008_df = race_2008_df.rename(columns=new_column_names_2008)

race_2008_df.head()
```

```python
# Save the 2008 race info to a CSV file
race_2008_df.to_csv('../CSV_Outputs/Race_2008_Data.csv', index=False)
```

#### Clean race census data in preparation for machine learning model

```python
# Read in demographic data race file
age_df = pd.read_csv("../CSV_Outputs/age_demo_data.csv",)

# Display sample data
age_df.head(5)
```

```python
# Get unique age groups
unique_age_groups = age_df['Age Group'].unique()

# Create new columns for each age group
for age_group in unique_age_groups:
    age_df[age_group] = (age_df['Age Group'] == age_group).astype(int)

# Drop the original 'Age Group' column if it's no longer needed
age_df = age_df.drop(columns=['Age Group'])
```

```python
age_df.head()
```

```python
# Drop columns that contain only NaN values
age_df = age_df.dropna(axis=1, how='all')

age_df.head()
```

```python
# Save the age census info to a CSV file
age_df.to_csv('../CSV_Outputs/Merged_Age_Data.csv', index=False)
```
