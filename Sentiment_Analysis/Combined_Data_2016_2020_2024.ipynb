```python
# import dependencies

import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Merging of the detailed CSVs which contain individual posts, their corresponding positive, negative and neutral sentiment scores, and the candidate the sentiment refers to.

##### 2016 Dataset cleaning and preparation for merge

```python
# Read in 2016 CSV file
ind_post_data_2016_df = pd.read_csv("CSV_Output_Sentiment/sent_and_cand_by_post_2016.csv",)

# Display sample data
ind_post_data_2016_df.head(5)
```

```python
# Drop columns that start with 'token_'
columns_to_drop = [col for col in ind_post_data_2016_df.columns if col.startswith('token_')]
ind_post_data_2016_df = ind_post_data_2016_df.drop(columns=columns_to_drop)

# Verify the result
ind_post_data_2016_df.head()
```

```python
# Add a column 'Year' with a date value for each row
ind_post_data_2016_df['Year'] = 2016

# Verify the result
ind_post_data_2016_df.head()
```

```python
# Rename the columns
ind_post_data_2016_df = ind_post_data_2016_df.rename(columns={
    'text': 'Text',
    'candidate': 'Candidate'
})

# Display the updated DataFrame
ind_post_data_2016_df.head()
```

##### 2020 Dataset cleaning and preparation for merge

```python
# Read in 2020 CSV file
ind_post_data_2020_df = pd.read_csv("CSV_Output_Sentiment/sent_and_cand_by_post_2020.csv",)

# Display sample data
ind_post_data_2020_df.head(5)
```

```python
# Define the desired order of columns
desired_columns = ['Compound', 'Positive', 'Negative', 'Neutral', 'text', 'candidate']

# Select and reorder the DataFrame columns
ind_post_data_2020_df = ind_post_data_2020_df[desired_columns]

# Display the resulting DataFrame
ind_post_data_2020_df.head()
```

```python
# Add a column 'Year' with a date value for each row
ind_post_data_2020_df['Year'] = 2020

# Verify the result
ind_post_data_2020_df.head()
```

```python
# Rename the columns
ind_post_data_2020_df = ind_post_data_2020_df.rename(columns={
    'text': 'Text',
    'candidate': 'Candidate'
})

# Display the updated DataFrame
ind_post_data_2020_df.head()
```

```python
# Replace values in the 'Candidate' column
ind_post_data_2020_df['Candidate'] = ind_post_data_2020_df['Candidate'].replace({
    'democrat': 'Democrat',
    'republican': 'Republican'
})

# Display the updated DataFrame
ind_post_data_2020_df.head()
```

##### 2024 Dataset cleaning and preparation for merge

```python
# Read in 2024 CSV file
ind_post_data_2024_df = pd.read_csv("CSV_Output_Sentiment/sent_and_cand_by_post_2024.csv",)

# Display sample data
ind_post_data_2024_df.head(5)
```

```python
# Drop columns that start with 'token_'
columns_to_drop = [col for col in ind_post_data_2024_df.columns if col.startswith('token_')]
ind_post_data_2024_df = ind_post_data_2024_df.drop(columns=columns_to_drop)

# Filter out rows where the 'Candidate' column is not 'Republican' or 'Democrat'
valid_candidates = ['republican', 'democrat']
ind_post_data_2024_df = ind_post_data_2024_df[ind_post_data_2024_df['candidate'].isin(valid_candidates)]

# Verify the result
ind_post_data_2024_df.head()
```

```python
# Add a column 'Year' with a date value for each row
ind_post_data_2024_df['Year'] = 2024

# Verify the result
ind_post_data_2024_df.head()
```

```python
# Rename the columns
ind_post_data_2024_df = ind_post_data_2024_df.rename(columns={
    'text': 'Text',
    'candidate': 'Candidate'
})

# Display the updated DataFrame
ind_post_data_2024_df.head()
```

```python
# Replace values in the 'Candidate' column
ind_post_data_2024_df['Candidate'] = ind_post_data_2024_df['Candidate'].replace({
    'democrat': 'Democrat',
    'republican': 'Republican'
})

# Display the updated DataFrame
ind_post_data_2024_df.head()
```

##### Concatenate the 3 dataframes into a single dataframe

```python
# Concatenate them into a single DataFrame
final_ind_post_data_df = pd.concat([ind_post_data_2016_df, ind_post_data_2020_df, ind_post_data_2024_df], ignore_index=True)

# Verify the result
print(f"Total rows in combined DataFrame: {len(final_ind_post_data_df)}")
final_ind_post_data_df.head()
```

```python
# Save the final combined individual post data dataFrame to a CSV file
final_ind_post_data_df.to_csv('../CSV_Outputs/final_ind_post_data_df.csv', index=False)
```

```python
# pie chart group visualizations

# Filter the DataFrame for Democrat and Republican candidates
dem_df = final_ind_post_data_df[final_ind_post_data_df['Candidate'] == 'Democrat']
rep_df = final_ind_post_data_df[final_ind_post_data_df['Candidate'] == 'Republican']
```

```python
# Calculate average sentiment scores for each year
years = [2016, 2020, 2024]
average_sentiments_dem = {}

for year in years:
    year_dem_df = dem_df[dem_df['Year'] == year]
    avg_positive_dem = year_dem_df['Positive'].mean()
    avg_negative_dem = year_dem_df['Negative'].mean()
    avg_neutral_dem = year_dem_df['Neutral'].mean()
    average_sentiments_dem[year] = {
        'Positive': avg_positive_dem,
        'Negative': avg_negative_dem,
        'Neutral': avg_neutral_dem
    }
```

```python
# Create pie charts
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for ax, year in zip(axes, years):
    sentiments = average_sentiments_dem[year]
    labels = sentiments.keys()
    sizes = sentiments.values()
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'lightblue', 'lightgrey'])
    ax.set_title(f'Democratic Sentiment Distribution {year}')

# Adjust layout
plt.tight_layout()
plt.savefig('../Visualizations/Democratic_Sentiment_By_Year.png')

# Display the pie charts
plt.show()
```

```python
# Calculate average sentiment scores for each year
years = [2016, 2020, 2024]
average_sentiments_rep = {}

for year in years:
    year_df = rep_df[rep_df['Year'] == year]
    avg_positive = year_df['Positive'].mean()
    avg_negative = year_df['Negative'].mean()
    avg_neutral = year_df['Neutral'].mean()
    average_sentiments_rep[year] = {
        'Positive': avg_positive,
        'Negative': avg_negative,
        'Neutral': avg_neutral
    }
```

```python
# Create pie charts
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for ax, year in zip(axes, years):
    sentiments = average_sentiments_rep[year]
    labels = sentiments.keys()
    sizes = sentiments.values()
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['red', 'lightcoral', 'lightgrey'])
    ax.set_title(f'Republican Sentiment Distribution {year}')

# Adjust layout
plt.tight_layout()
plt.savefig('../Visualizations/Republican_Sentiment_By_Year.png')

# Display the pie charts
plt.show()
```
