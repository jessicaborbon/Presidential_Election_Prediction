```python
# For comma-separated values
race_2008_df = pd.read_csv('../censusoutputcsv/2008_census_race.csv', sep=',')
race_2012_df = pd.read_csv('../censusoutputcsv/2012_census_race.csv', sep=',')
race_2016_df = pd.read_csv('../censusoutputcsv/2016_census_race.csv', sep=',')
race_2020_df = pd.read_csv('../censusoutputcsv/2020_census_race1.csv', sep=',')
```

```python
# List of demographic percentage columns to convert to fractions for averaging
demographic_percentage_columns = [
    'white',
    'black_or_african_american',
    'american_indian_and_alaska_native',
    'asian',
    'native_hawaiian_and_other_pacific-islander',
    'hispanic_or_latino_any_race'
]

# Convert the demographic percentage columns to fractions by dividing by 100
race_2008_df[demographic_percentage_columns] = race_2008_df[demographic_percentage_columns] / 100

# Group by 'state' and calculate the average of all numerical columns
state_averages2008_df = race_2008_df.groupby('state').mean().reset_index()

# Convert the specific percentage columns back to percentages by multiplying by 100
state_averages2008_df[demographic_percentage_columns] = state_averages2008_df[demographic_percentage_columns] * 100

# Remove percentage formatting step (do not add '%' symbol)
# Format the specific percentage columns to show as plain numbers with one decimal place
for col in demographic_percentage_columns:
    state_averages2008_df[col] = state_averages2008_df[col].map(lambda x: round(x, 1))

# Convert year to whole integer
state_averages2008_df['year'] = state_averages2008_df['year'].astype(int)

# Display the results
state_averages2008_df.head()
```

```python
# Rename columns to add a _2008 suffix
state_averages2008_df.rename(columns={
    'white': 'white_2008',
    'black_or_african_american': 'black_or_african_american_2008',
    'american_indian_and_alaska_native': 'american_indian_and_alaska_native_2008',
    'asian': 'asian_2008',
    'native_hawaiian_and_other_pacific-islander': 'native_hawaiian_and_other_pacific-islander_2008',
    'hispanic_or_latino_any_race': 'hispanic_or_latino_any_race_2008'
}, inplace=True)

# Display the results
state_averages2008_df.head()
```

```python
# Merge the DataFrames on 'state' (outer join to include all states)
merged_df = state_averages2008_df.merge(state_averages2012_df, on='state', how='outer', suffixes=('', '_2012'))
merged_df = merged_df.merge(state_averages2016_df, on='state', how='outer', suffixes=('', '_2016'))
merged_df = merged_df.merge(state_averages2020_df, on='state', how='outer', suffixes=('', '_2020'))

# Handle any duplicate columns caused by the merges
merged_df = merged_df.loc[:, ~merged_df.columns.duplicated()]

# Export the merged DataFrame to a CSV file
merged_df.to_csv('merged_race_data.csv', index=False)

# Display the merged DataFrame
merged_df.head()
```
