```python
# List of years to filter by
years_to_include = [2008, 2012, 2016, 2020, 2022, 2024]

# Filter the DataFrame by the specified years
df_election = df3[df3['year'].isin(years_to_include)]
```

```python
 df_selected = df_election[['year', 'winner']]
```

```python
df_selected
```

```python
df_cleaned = df_selected.dropna()
```

```python
df_cleaned
```

```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder \
    .appName("PySparkJupyter") \
    .config("spark.driver.memory", "2g") \
    .getOrCreate()

# Check Spark session details
spark
```

```python
# Convert pandas DataFrames to PySpark DataFrames
large_spark_df = spark.createDataFrame(df1)
small_spark_df = spark.createDataFrame(df_cleaned)
```

```python
# Perform the join (merge)
merged_spark_df = large_spark_df.join(small_spark_df, on="year", how="inner")

# Show the merged DataFrame
merged_spark_df.show(5)
```

```python
# Show the column names
column_names = merged_spark_df.columns
print("Columns in the DataFrame:")
for name in column_names:
    print(name)
```

```python
# Rename conflicting columns
from pyspark.sql.functions import col

merged_spark_df = merged_spark_df.withColumnRenamed("candidate", "candidate_2") 
```

```python
# Show the column names
column_names = merged_spark_df.columns
print("Columns in the DataFrame:")
for name in column_names:
    print(name)
# Rename columns with duplicates
new_column_names = []
seen_columns = {}

for name in column_names:
    if name in seen_columns:
        # Create a unique name by appending a counter suffix
        count = seen_columns[name]
        new_name = f"{name}_{count + 1}"
        seen_columns[name] += 1
    else:
        new_name = name
        seen_columns[name] = 1
    new_column_names.append(new_name)

# Create a DataFrame with renamed columns
for old_name, new_name in zip(column_names, new_column_names):
    if old_name != new_name:
        merged_spark_df = merged_spark_df.withColumnRenamed(old_name, new_name)
```
