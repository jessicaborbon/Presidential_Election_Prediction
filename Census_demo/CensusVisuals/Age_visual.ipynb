```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'reduced_age_census.csv'  # Replace with your file path
df = pd.read_csv(file_path)

# Filter the data for the years of interest
years_of_interest = [2008, 2012, 2016, 2020, 2022]
filtered_df = df[df['year'].isin(years_of_interest)]

# Group by year and calculate the average median age across all states
median_age_by_year = filtered_df.groupby('year')['median_age(years)'].mean()

# Plot the median age change over the selected years
plt.figure(figsize=(10, 6))
plt.plot(median_age_by_year.index, median_age_by_year.values, marker='o', linestyle='-', color='b')
plt.title('Average Median Age Change (2008-2022)')
plt.xlabel('Year')
plt.ylabel('Average Median Age (Years)')
plt.grid(True)
plt.show()
```

```python
# Save the plot as a PNG file
plt.savefig('median_age_change_2008_2022.png')

# Display the plot
plt.show()
```
